Calculate the royalty of each sale for each author.

```python
query_1='''

SELECT
        t.title_id as title_id
        , a.au_id as author_id
        , (t.price * s.qty * t.royalty/100 * ta.royaltyper / 100) AS royalties
    FROM 
        `ironhack-data-analytics-265219.publications.titles` t
    INNER JOIN
        `ironhack-data-analytics-265219.publications.sales` s 
    ON
        s.title_id = t.title_id
    INNER JOIN
        `ironhack-data-analytics-265219.publications.titleauthor` ta 
    ON
        ta.title_id = s.title_id
    INNER JOIN
        `ironhack-data-analytics-265219.publications.authors` a 
    ON
        a.au_id = ta.au_id
    
    ORDER BY 1
        '''
```

```python
query_job_1 = client.query(query=query_1)
df_1=query_job_1.to_dataframe()
df_1
```

```python
query_2='''

SELECT
    title_id
    , au_id as author_id
    , au_lname
    , au_fname
    , advance
    , SUM(royalties) AS royalties 
FROM 
    (SELECT
        t.title_id
        , t.price
        , t.advance
        , t.royalty
        , s.qty
        , a.au_id
        , au_lname
        , au_fname
        , ta.royaltyper
        , (t.price * s.qty * t.royalty/100 * ta.royaltyper / 100) AS royalties
    FROM 
        `ironhack-data-analytics-265219.publications.titles` t
    INNER JOIN
        `ironhack-data-analytics-265219.publications.sales` s 
    ON
        s.title_id = t.title_id
    INNER JOIN
        `ironhack-data-analytics-265219.publications.titleauthor` ta 
    ON
        ta.title_id = s.title_id
    INNER JOIN
        `ironhack-data-analytics-265219.publications.authors` a 
    ON
        a.au_id = ta.au_id)a
GROUP BY
    1,2,3,4,5'''
```

```python
query_job_2 = client.query(query=query_2)
df_2=query_job_2.to_dataframe()
df_2
```

```python
query_3='''
WITH royalties_table AS(
SELECT
    title_id
    , au_id
    , au_lname
    , au_fname
    , advance
    , SUM(royalties) AS royalties 
FROM 
    (SELECT
        t.title_id
        , t.price
        , t.advance
        , t.royalty
        , s.qty
        , a.au_id
        , au_lname
        , au_fname
        , ta.royaltyper
        , (t.price * s.qty * t.royalty/100 * ta.royaltyper / 100) AS royalties
    FROM 
        `ironhack-data-analytics-265219.publications.titles` t
    INNER JOIN
        `ironhack-data-analytics-265219.publications.sales` s 
    ON
        s.title_id = t.title_id
    INNER JOIN
        `ironhack-data-analytics-265219.publications.titleauthor` ta 
    ON
        ta.title_id = s.title_id
    INNER JOIN
        `ironhack-data-analytics-265219.publications.authors` a 
    ON
        a.au_id = ta.au_id)a
GROUP BY
    1,2,3,4,5)
            SELECT
                au_id AS author_id
                , au_lname AS last_name
                , au_fname AS first_name
                , sum(advance + royalties) AS profits
            FROM
                royalties_table
            GROUP BY
                1,2,3
            ORDER BY
                profits DESC
            LIMIT 3
'''
```

```python
query_job_3 = client.query(query=query_3)
df_3=query_job_3.to_dataframe()
df_3
```
